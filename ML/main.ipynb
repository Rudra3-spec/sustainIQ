```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
import joblib

# 1. Generate High-Variety Synthetic Data
np.random.seed(42)
n_samples = 1000
data = {
    "building_density": np.random.uniform(10, 100, n_samples),
    "road_connectivity": np.random.uniform(1, 10, n_samples),
    "public_transit_access": np.random.uniform(1, 100, n_samples),
    "air_quality_index": np.random.uniform(20, 500, n_samples),
    "green_cover_percentage": np.random.uniform(5, 50, n_samples),
    "carbon_footprint": np.random.uniform(10, 100, n_samples),
    "population_density": np.random.uniform(100, 5000, n_samples),
    "crime_rate": np.random.uniform(1, 100, n_samples),
    "avg_income": np.random.uniform(10, 200, n_samples),
    "renewable_energy_usage": np.random.uniform(0, 1, n_samples),
    "disaster_risk_index": np.random.uniform(1, 100, n_samples)
}

df = pd.DataFrame(data)

# 2. Create a realistic target score (Simulation)
# Green cover and Transit UP = Score UP; Crime and Carbon UP = Score DOWN
df['sustainability_score'] = (
    (df['green_cover_percentage'] * 0.3) + 
    (df['public_transit_access'] * 0.2) - 
    (df['carbon_footprint'] * 0.2) - 
    (df['crime_rate'] * 0.1)
)
# Normalize score between 0 and 1
df['sustainability_score'] = (df['sustainability_score'] - df['sustainability_score'].min()) / (df['sustainability_score'].max() - df['sustainability_score'].min())

# 3. Train and Save
X = df.drop('sustainability_score', axis=1)
y = df['sustainability_score']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_scaled, y)

joblib.dump(model, 'sustainability_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("✅ New high-sensitivity model and scaler saved!")
```
